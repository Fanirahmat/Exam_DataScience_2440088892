```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('rollingsales_brooklyn.csv')
data
```

```python
# Visualisasi 1: Histogram Harga Properti
plt.figure(figsize=(10, 6))
plt.hist(data['sale.price'], bins=20, edgecolor='black')
plt.title("Distribusi Harga Properti")
plt.xlabel("Harga Properti")
plt.ylabel("Frekuensi")
plt.show()
```

```python
# Visualisasi 2: Grafik Scatter antara Luas Tanah dan Harga Properti
plt.figure(figsize=(10, 6))
plt.scatter(data['land.square.feet'], data['sale.price'], alpha=0.5)
plt.title("Hubungan antara Luas Tanah dan Harga Properti")
plt.xlabel("Luas Tanah (sqft)")
plt.ylabel("Harga Properti")
plt.show()
```

```python
# Visualisasi 3: Grafik Batang Top 10 Neighborhood dengan Jumlah Penjualan Properti Terbanyak
top_10_neighborhood = data['neighborhood'].value_counts().head(10)
plt.figure(figsize=(10, 6))
top_10_neighborhood.plot(kind='bar')
plt.title("Top 10 Neighborhood dengan Jumlah Penjualan Properti Terbanyak")
plt.xlabel("Neighborhood")
plt.ylabel("Jumlah Penjualan")
plt.xticks(rotation=45)
plt.show()
```
